L'obiettivo di questa lezione era scrivere un proprio algoritmo genetico per risolvere il problema del commesso viaggiatore.

Ho cercato di seguire il più possibile le indicazioni fornite nel jupyter notebook di introduzione all'esercitazione:

- Il percorso è espresso da un vettore monodimensionale;
- Una check-function verifica che ogni individuo della popolazione soddisfi i vincoli del problema;
- La popolazione è ordinata in modo da fittare la funzione $L^{(1)}(\overline{x_1},\overline{x_2},..., \overline{x_N})$
- L'operatore di selezione è anch'esso quello suggerito a lezione, ossia partendo dalla popolazione ordinata sceglie l'individuo casualmente dando maggior probabilità a quello che fitta meglio;
- I cinque operatori di mutazione genetica sono anch'essi quelli descritti nel jupyter notebook ( con probabilità $p_m$)  così come l'operatore crossover (con probabilità $p_c$).

Per generare le disposizioni delle città su una circonferenza o entro un quadrato ho realizzato uno script python che le scrive rispettivamente nei file *circle_world.config* e *square_world.config*. Il codice C++ richiede in input la configurazione delle città dal file *world_config.dat* e un file *input.dat* per gli altri parametri.

Col codice implementato ho quindi ottimizzato il path di due percorsi di 32 città generate a caso su una circonferenza o in un quadrato.

Ho scelto come parametri:

- 100 cromosomi;
- 2000 generazioni dell'algoritmo;
- Esponente nell'operatore di selezione $p=2$;
- Probabilità dell'operatore crossover $p_c=80\%$;
- Probabilità mutazioni $p_m=50\%$;

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carico i dati del mondo quadrato
x_sq_avg, y_sq_avg = np.loadtxt("../results/square/average_distance_over_best_half_paths.dat", unpack = True)
x_sq_best, y_sq_best = np.loadtxt("../results/square/distance_of_best_path.dat", unpack = True)
x_sq_path, y_sq_path = np.loadtxt("../results/square/best_path.dat", unpack = True)
# Carico i dati del mondo circolare
x_circ_avg, y_circ_avg = np.loadtxt("../results/circle/average_distance_over_best_half_paths.dat", unpack = True)
x_circ_best, y_circ_best = np.loadtxt("../results/circle/distance_of_best_path.dat", unpack = True)
x_circ_path, y_circ_path = np.loadtxt("../results/circle/best_path.dat", unpack = True)

# Plotto le distanze L1 migliori e medie per ogni generazione (square world)
plt.figure(1, figsize = (14, 7))
plt.title("Fitness value = $L^{(1)}$ distance (square world)")
plt.plot(x_sq_best, y_sq_best, label="Best fitness value $L^{(1)}$")
plt.plot(x_sq_avg, y_sq_avg, label="Average fitness value $< L^{(1)}>$")
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.grid(True)
plt.legend()

# Plotto le distanze L1 migliori e medie per ogni generazione (circle world)
plt.figure(2, figsize = (14, 7))
plt.title("Fitness value = $L^{(1)}$ distance (circular world)")
plt.plot(x_circ_best, y_circ_best, label="Best fitness value $L^{(1)}$")
plt.plot(x_circ_avg, y_circ_avg, label="Average fitness value $< L^{(1)}>$")
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.grid(True)
plt.legend()

# Plotto il percorso migliore
plt.figure(3, figsize = (14, 7))

# Square world
plt.subplot(1, 2, 1)
plt.title("Best path (square world)")
plt.plot(x_sq_path, y_sq_path,  marker=".")
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((-1.2, 1.2))
plt.ylim((-1.2, 1.2))
plt.xticks([])
plt.yticks([])
# Circle world
plt.subplot(1, 2, 2)
plt.title("Best path (circular world)")
plt.plot(x_circ_path, y_circ_path,  marker=".")
plt.xlabel('x')
plt.ylabel('y')

plt.xlim((-1.2, 1.2))
plt.ylim((-1.2, 1.2))
plt.xticks([])
plt.yticks([])

plt.show()
```
